# Data Cleaning

```python
# Splitting the 'Disease' column into lists
df['Disease'] = df['Disease'].str.split(', ')

# Counting the frequency of each disease
disease_counts = pd.Series([disease for diseases in df['Disease'] for disease in diseases]).value_counts()

# Displaying the frequency of diseases
display(disease_counts)

# Getting the list of unique diseases
common_diseases = disease_counts.index
```

```python
for disease in common_diseases:
    df[disease] = df['Disease'].apply(lambda diseases: 1 if disease in diseases else 0)

# Displaying the updated DataFrame
display(df.head())
```

# Feature Engineering

```python
# Manually mapping 

activity_mapping = {
    'Sedentary': 0,
    'Lightly Active': 1,
    'Moderately Active': 2,
    'Very Active': 3,
    'Extremely Active': 4
}

df['Activity Level'] = df['Activity Level'].map(activity_mapping)

# One-Hot Encoding for 'Dietary Preference'
df = pd.get_dummies(df, columns=['Dietary Preference'])

# Converting One-Hot Encoded columns to integers
for col in df.columns:
    if 'Dietary Preference' in col:
        df[col] = df[col].astype(int)

# Displaying the final DataFrame
df
```

# Data Normalization

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# List of feature columns to use for normalization
feature_columns = [
    'Ages', 'Weight', 'Activity Level',
    'Weight Gain', 'Hypertension', 'Heart Disease', 'Kidney Disease',
    'Diabetes', 'Acne', 'Weight Loss',
    'Dietary Preference_Omnivore', 'Dietary Preference_Pescatarian',
    'Dietary Preference_Vegan', 'Dietary Preference_Vegetarian'
]

# Function to normalize both meal features and user profile
def normalize_data(df, user_profile, feature_columns):
    scaler = StandardScaler()

    # Normalizing meal features
    meal_features_normalized = scaler.fit_transform(df[feature_columns])

    # Normalizing user profile
    user_profile_vector = pd.DataFrame([user_profile])[feature_columns]
    user_profile_normalized = scaler.transform(user_profile_vector)

    return meal_features_normalized, user_profile_normalized
```

# Content-Based Recommendation System

```python
# Function to recommend meals
def recommend_meals_content_based(user_profile_normalized, meal_features_normalized, df):
    # Calculating similarity
    similarity_scores = cosine_similarity(user_profile_normalized, meal_features_normalized)

    # Copying df to avoid modifying the original dataset
    df_copy = df.copy()
    df_copy['Similarity'] = similarity_scores[0]

    # Recommend top 5 meals
    recommendations = df_copy.sort_values(by='Similarity', ascending=False).head(5)

    return recommendations[
    ['Breakfast Suggestion', 'Lunch Suggestion', 'Dinner Suggestion', 'Snack Suggestion',
     'Protein', 'Sugar', 'Sodium', 'Calories', 'Carbohydrates', 'Fiber', 'Fat', 'Similarity']
]
```

```python
# Defining only the selected features (others will be 0 by default)
selected_features = { 
    'Ages': 65,
    'Weight': 80,
    'Activity Level': 2, # lightly active
    'Weight Gain': 1,
    'Dietary Preference_Omnivore': 1
}

# Creating a user profile with all features set to 0 by default
user_profile = {feature: 0 for feature in feature_columns}

# Updating only the selected features
user_profile.update(selected_features)

# Normalizing data
meal_features_normalized, user_profile_normalized = normalize_data(df, user_profile, feature_columns)

# Get top 5 recommended meals
recommendations = recommend_meals_content_based(user_profile_normalized, meal_features_normalized, df)

# Print recommendations
recommendations
```
